```python
import os
import numpy as np
import pandas as pd
```

```python
def drop_columns(df, cols):
    """
    This function removes the columns from a DataFrame.
    """
    for col in cols:
       try:
              df.drop(col, axis=1, inplace=True)
       except Exception as e:
              print(f"Bir hata oluştu: {e}")
```

```python
replace_dict = {
    'Ara Kat': '3. Kat',
    'En Üst Kat': '5. Kat',
    'Bahçe Katı': '0. Kat',
    'Yüksek Giriş': '1. Kat',
    'Yarı Bodrum': '-1. Kat',
    'Giriş Katı': '0. Kat',
    'Çatı Katı': '5. Kat',
    'Villa Katı': '0. Kat',
    'Zemin': '0. Kat',
    'Kot 1': '-1. Kat',
    'Kot 2': '-2. Kat',
    'Kot 3': '-3. Kat',
    '21 ve üzeri': '21. Kat',
    'Bodrum ve Zemin': '0. Kat',
    'Asma Kat': '1. Kat',
    'Tripleks': '0. Kat',
    'Teras Katı': '5. Kat',
    'Bodrum': '-1. Kat',
    'nan': '2. Kat'
}

df['celly 4'] = df['celly 4'].replace(replace_dict.keys(), replace_dict.values()).astype(str)
df['celly 4'].dropna(inplace=True)
df['floor'] = df['celly 4'].apply(lambda x: x.split('.')[0]).astype(int)
```
